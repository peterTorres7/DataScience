```python
# imports
import pandas as pd
import requests
from bs4 import BeautifulSoup

# get data
url = 'https://www.worldometers.info/coronavirus/country/us/#nav-yesterday'
response = requests.get(url)
data = response.text

# parse data and get table
soup = BeautifulSoup(data, 'html.parser')
table = soup.find(id="usa_table_countries_today")

# exclude #, source, projections, totals
df = pd.read_html(str(table))[0]
df.drop(['#', 'Source', 'Projections'], axis=1, inplace=True)
df.drop([0, 64], axis=0, inplace=True)

# first few rows
print("First few rows")
print(df.head(10))

# summary statistics
print("\nSummary statistics")
print(df.describe())

# top 5 cases
print("\nTop 5 new cases")
print(df[["USAState", "NewCases"]].sort_values(by=["NewCases"], ascending=False).head())

# top 5 deaths
print("\nTop 5 total deaths")
print(df[["USAState", "TotalDeaths"]].sort_values(by=["TotalDeaths"], ascending=False).head())

# top 5 cases / 1M pop
print("\nTop 5 total cases / 1M pop")
total_cases = df.iloc[:, [0, 7]]
print(total_cases.sort_values(by=[total_cases.columns[1]], ascending=False).head())

# top 5 deaths / 1M pop
print("\nTop 5 total deaths / 1M pop")
total_deaths = df.iloc[:, [0, 8]]
print(total_deaths.sort_values(by=[total_deaths.columns[1]], ascending=False).head())
```
